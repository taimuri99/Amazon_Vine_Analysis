```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("M16-Amazon-Challenge").config("spark.driver.extraClassPath","/content/postgresql-42.2.16.jar").getOrCreate()
```

```python
from pyspark import SparkFiles
url = "https://s3.amazonaws.com/amazon-reviews-pds/tsv/amazon_reviews_us_Sports_v1_00.tsv.gz"
spark.sparkContext.addFile(url)
df = spark.read.option("encoding", "UTF-8").csv(SparkFiles.get("amazon_reviews_us_Sports_v1_00.tsv.gz"), sep="\t", header=True, inferSchema=True)
df.show()
```

```python
# Create the vine_table. DataFrame
vine_df = df.select(['review_id', 'star_rating', 'helpful_votes', 'total_votes', 'vine', 'verified_purchase'])
vine_df.show()
# change star_rating to int
from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType
vine_df = vine_df.withColumn('star_rating',col('star_rating').cast(IntegerType()))
vine_df.dtypes
```

```python
# Filter the data and create a new DataFrame or table to retrieve all the rows where the total_votes count is equal to or greater than 20 to pick reviews that are more likely to be helpful and to avoid having division by zero errors later on.
df1 = vine_df.filter(vine_df['total_votes']>=20)
df1.show()
```

```python
# Filter the new DataFrame or table created in Step 1 and create a new DataFrame or table to retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.
df2 = df1.filter((df1['helpful_votes']/df1['total_votes']>=0.5))
df2.show()
```

```python
#Filter the DataFrame or table created in Step 2, and create a new DataFrame or table that retrieves all the rows where a review was written as part of the Vine program (paid), vine == 'Y'.
# PAID
df3 = df2.filter(df2['vine']=='Y')
df3.show()
```

```python
# same as df3 but with vine == N
# UNPAID
df4 = df2.filter(df2['vine']=='N')
df4.show()
```

```python
# PAID
# total paid reviews
total_paid_reviews = df3.count()
total_paid_reviews
```

```python
#total paid 5 star reviews
total_paid_five = df3.filter(df3['star_rating']==5).count()
total_paid_five
```

```python
# percentage of 5-star reviews PAID
perc_five_paid = (total_paid_five/total_paid_reviews)*100
perc_five_paid
```

```python
# UNPAID
# total unpaid reviews
total_unpaid_reviews = df4.count()
total_unpaid_reviews
```

```python
#total unpaid 5 star reviews
total_unpaid_five = df4.filter(df4['star_rating']==5).count()
total_unpaid_five
```

```python
# percentage of 5-star reviews PAID
perc_five_unpaid = (total_unpaid_five/total_unpaid_reviews)*100
perc_five_unpaid
```

```python
print(f'Total number of Vine reviews: {total_paid_reviews}')
print(f'Total number of Vine reviews with 5 stars: {total_paid_five}')
print(f'Percentage of 5 star Vine reviews: {round(perc_five_paid)}%')

print(f'Total number of non Vine reviews: {total_unpaid_reviews}')
print(f'Total number of 5 star non Vine reviews: {total_unpaid_five}')
print(f'Percentage of 5 star non Vine reviews: {round(perc_five_unpaid)}%')
```
